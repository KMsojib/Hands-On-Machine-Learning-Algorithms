*** Dot Products ***


```python
a = np.array([[1,2],[3,4]]) 
b = np.array([[11,12],[13,14]]) 
print("Dot Product np.dot(a,b)):\n", np.dot(a,b))

a = np.array([[1,2],[3,4]])
b = np.array([[11,12],[13,14]])

b_transpose = np.transpose(b)
result = np.dot(a,b_transpose)
print("Dot  Product np.dot(a,b_transpose)):\n", result)
```

```python
a = np.array([[1],[2]])
a_t = np.transpose(a)

w = np.array([[3,4],[5,6]])  # Corrected W definition

z = np.dot(a_t, w)
print("Dot Product np.dot(a_t, w)):\n", z)
```

```python
a = np.array([[1, -1], [2, -2]])
w = np.array([[3, 4], [5, 6]])

a_t = np.transpose(a)

z = np.dot(a_t, w)

print("Matrix a:")
print(a)

print("\nTransposed matrix a_t:")
print(a_t)

print("\nMatrix w:")
print(w)

print("\nDot Product np.dot(a_t, w):")
print(z)
```

```python
import numpy as np

a = np.array([[200,17]])
a_t = np.transpose(a)
w = np.array([[1, -3, 5], [-2, 4, -6]])
b = np.array([[-1, 1, 2]]) 

def dense(a, W, b, g=lambda x: x): # Changed AT to a
    z = np.matmul(a, W) + b  # use a instead of a_t
    a_out = g(z)
    return a_out

output_identity = dense(a, w, b) # Changed a_t to a
print("Output with identity activation:\n", output_identity)

def sigmoid(x):
    return 1/(1+np.exp(-x))

output_sigmoid = dense(a, w, b, g=sigmoid) # Changed a_t to a
print("\nOutput with sigmoid activation:\n", output_sigmoid)
```
