```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Prepare Data (example with a hypothetical dataset)
data = {
    'Age': np.random.randint(18, 60, 100),
    'Income': np.random.normal(50000, 15000, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Education': np.random.choice(['High School', 'Bachelors', 'Masters', 'PhD'], 100)
}
df = pd.DataFrame(data)

# Create a figure with two subplots
plt.figure(figsize=(12, 5))

# Subplot 1: Scatterplot
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
sns.scatterplot(x='Age', y='Income', hue='Gender', data=df, s=100, alpha=0.7)
plt.title("Scatter Plot: Age vs. Income by Gender")
plt.xlabel("Age")
plt.ylabel("Income")

# Subplot 2: Countplot
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
sns.countplot(x='Education', data=df, palette='viridis', order=df['Education'].value_counts().index)
plt.title("Count Plot: Distribution of Education Levels")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

plt.tight_layout() # Adjust layout
plt.show()
```
