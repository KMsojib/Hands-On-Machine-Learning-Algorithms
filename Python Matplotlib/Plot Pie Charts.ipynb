```python
import matplotlib.pyplot as plt

# 1. Prepare Data (e.g., market share of different products)
sizes = [30, 25, 20, 15, 10]
labels = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']
explode = (0, 0.1, 0, 0, 0)  # Explode 2nd slice

# 2. Create Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

# 3. Add Title
plt.title("Market Share Distribution")

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# 4. Display Plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare Data (e.g., ages of a group of people)
ages = np.random.normal(loc=30, scale=10, size=1000) # Normally distributed ages

# 2. Create Plot
plt.hist(ages, bins=20, edgecolor='black') # bins define the number of intervals

# 3. Add Labels and Title
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Ages")

# 4. Display Plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.exponential(scale=2, size=500) # Exponentially distributed data

plt.hist(data, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='darkblue', label='Data Distribution')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Histogram with Density")
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```
