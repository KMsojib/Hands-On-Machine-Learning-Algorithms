```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D # Important: import this for 3D projections

# 1. Prepare 3D Data
np.random.seed(42)
num_points = 100
x = np.random.rand(num_points) * 10
y = np.random.rand(num_points) * 10
z = np.random.rand(num_points) * 10

# 2. Create Figure and 3D Axes
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d') # '111' means 1x1 grid, 1st subplot

# 3. Create 3D Scatter Plot
ax.scatter(x, y, z, c='skyblue', marker='o', s=50)

# 4. Add Labels and Title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("Basic 3D Scatter Plot")

# 5. Display Plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Prepare data for a surface plot (e.g., a "Mexican Hat" function)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y) # Create a grid of x, y values
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) / R # Mexican Hat function

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create a 3D surface plot
surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3D Surface Plot of sin(R)/R")
fig.colorbar(surface, shrink=0.5, aspect=5, label='Z Value') # Add a color bar
plt.show()
```
