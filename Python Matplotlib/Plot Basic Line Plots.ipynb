# Plot Basic Line Plots

###### Line plots : are used to display data points connected by straight line segments. They are ideal for showing trends over time or ordered categories.

###### Concept: A line plot connects data points, typically showing the relationship between two variables where one is often dependent on the other.

##### Steps:
###### 1. Import Matplotlib: Bring in the matplotlib.pyplot module, commonly aliased as plt.
###### 2. Prepare Data: Create two lists or arrays for your ``` x and y ``` coordinates.
###### 3. Create Plot: Use ``` plt.plot() ``` to generate the line plot.
###### 4. Add Labels and Title: Use ``` plt.xlabel(), plt.ylabel(), and plt.title()``` for clarity.
###### 5. Display Plot: Use ```plt.show()``` to render the plot.

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
xPoint = np.array([0,6])
yPoint = np.array([0,250])

plt.plot(xPoint,yPoint)
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 6])

# 2. Create Plot
plt.plot(x, y, color = 'red', marker = 'o', linestyle = '-')

# 3. Add Labels and Title
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Basic Line Plot")

# 4. Display Plot
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # 100 points between 0 and 10
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='blue', linestyle='-', label='Sine Wave')
plt.plot(x, y2, color='red', linestyle='--', label='Cosine Wave')

plt.xlabel("Angle (radians)")
plt.ylabel("Amplitude")
plt.title("Multiple Line Plots")
plt.legend() # Show legend for labels
plt.grid(True) # Add a grid
plt.show()
````
