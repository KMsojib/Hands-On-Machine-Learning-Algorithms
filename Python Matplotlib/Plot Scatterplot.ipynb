```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare Data (e.g., relationship between study hours and exam scores)
study_hours = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
exam_scores = np.array([60, 65, 70, 75, 80, 85, 90, 92, 95])

# 2. Create Plot
plt.scatter(study_hours, exam_scores)

# 3. Add Labels and Title
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Scatter Plot: Study Hours vs. Exam Scores")

# 4. Display Plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42) # for reproducibility
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
colors = np.random.rand(50) # for color mapping
sizes = np.random.rand(50) * 500 # for size mapping

plt.scatter(x, y, c=colors, s=sizes, alpha=0.8, cmap='viridis') # cmap for color mapping
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Customized Scatter Plot")
plt.colorbar(label='Color Intensity') # Add a color bar
plt.show()
```
