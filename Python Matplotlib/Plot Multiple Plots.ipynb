```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Line Plot
x_line = np.linspace(0, 10, 50)
y_line = np.sin(x_line)

# Data for Scatter Plot
np.random.seed(0)
x_scatter = np.random.rand(30) * 10
y_scatter = np.random.rand(30) * 2 - 1 # between -1 and 1

# Plot Line
plt.plot(x_line, y_line, color='blue', label='Sine Wave')

# Plot Scatter
plt.scatter(x_scatter, y_scatter, color='red', marker='x', label='Random Points')

# Add Labels, Title, and Legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Plots on Single Axes")
plt.legend() # Display the labels defined in plot/scatter functions
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = np.exp(-x)

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) # figsize (width, height) in inches

# Plot 1: Top-Left (Line Plot)
axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title("Sine Wave")
axes[0, 0].set_xlabel("X")
axes[0, 0].set_ylabel("Y")

# Plot 2: Top-Right (Line Plot)
axes[0, 1].plot(x, y2, color='green', linestyle='--')
axes[0, 1].set_title("Cosine Wave")
axes[0, 1].set_xlabel("X")
axes[0, 1].set_ylabel("Y")

# Plot 3: Bottom-Left (Scatter Plot)
axes[1, 0].scatter(x[::10], y3[::10], color='red', marker='o') # Plot every 10th point
axes[1, 0].set_title("X Squared")
axes[1, 0].set_xlabel("X")
axes[1, 0].set_ylabel("Y")

# Plot 4: Bottom-Right (Line Plot)
axes[1, 1].plot(x, y4, color='purple', linewidth=2)
axes[1, 1].set_title("Exponential Decay")
axes[1, 1].set_xlabel("X")
axes[1, 1].set_ylabel("Y")

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the figure with subplots
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(12, 4))

# First subplot (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x, np.sin(x))
ax1.set_title("Sin(x)")

# Second subplot (1 row, 2 columns, second plot)
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(x, np.cos(x), color='orange')
ax2.set_title("Cos(x)")

plt.tight_layout()
plt.show()
```
